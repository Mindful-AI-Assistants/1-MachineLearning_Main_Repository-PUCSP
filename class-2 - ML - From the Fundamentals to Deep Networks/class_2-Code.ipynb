```python
def sigma(z):
  # delimitador rÃ­gido
  if z >= 0:
    return 1
  else:
    return 0

def sigmoid(z, a=1):
  return 1.0 / (1 + np.exp(-a*z))

def dif_sigmoid( z, a ):
    return sigmoid(z) * ( 1 - sigmoid(z) )
```

```python
z = w@x + b
a = sigma( z )
a
```

```python
z = 0

for i in range( len(x) ):
  z = z + w[i] * x[i]
z = z + b
print(z)

a = sigma(z)
print(a)
```

```python
sigmoid(z)
```

Graficos

```python
xx = np.linspace(-10,10, 100)

yy1 = np.array([sigma(x) for x in xx])

for a in [0.5, 0.75, 1, 10]:
  yy2 = sigmoid(xx, a)
  plt.plot(xx, yy1, xx, yy2)
```

```python
xx = np.linspace(-10,10, 100)

yy1 = np.array([sigma(x) for x in xx])

for a in [1]:
  yy2 = sigmoid(xx, a)
  dyy2 = dif_sigmoid(xx,a)
  plt.plot(xx, yy1, xx, yy2, xx, dyy2)
```
